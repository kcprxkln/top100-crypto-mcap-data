```python
import csv
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
with open ('mcap_data.csv', 'r') as f:
    reader = csv.reader(f)
    rows = list(reader)
    
```

```python
dates = []
values = []
```

```python
for row in rows:
    dates.append(row[0])
    values.append(row[1])
```

```python
import datetime as dt
```

```python
dates = [dt.datetime.strptime(d, '%Y%m%d').date() for d in dates]
values = [int(value) for value in values]
```

```python
print(values)
```

```python
yticks = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]

f = plt.figure()
f.set_figwidth(24)
f.set_figheight(10)

plt.title(label = 'market cap of top 100 cryptocurrencies over the time', fontdict={'fontsize': 22, 'family':'sans-serif'})
plt.grid()
plt.plot(dates, values, linestyle='-', color='#3C32F0')
plt.ylabel('Marketcap ($)', fontdict={'fontsize':16, 'family':'sans-serif'})
plt.xlabel('Year', fontdict={'fontsize':16, 'family':'sans-serif'})
plt.savefig('top100_mcap_chart.jpg', dpi=300)
```
